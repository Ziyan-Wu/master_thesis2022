# Date barplot

```python
def create_dfdate(df_pts):
    newdf = pd.DataFrame(df_pts, columns=['t'])
    newdf['date'] = newdf.t.str.extract(('(^\d{4}-\d{1,2})'), expand=True)
    newdf = newdf.groupby('date').agg('count')
    newdf = newdf.rename(columns={'t': 'cnt'})
    return newdf
```

```python
df03 = create_dfdate(df_pts03)
df03
```

```python
df02 = create_dfdate(df_pts02)
df02
```

```python
df01 = create_dfdate(df_pts01)
df01
```

```python
def concat_dfdate(*args):
    result = []
    result = pd.concat(*args, axis=1)
    result = result.fillna(0)
    return result.sort_index()
```

```python
df = concat_dfdate([df01,df02,df03])
df.columns = ['cnt01', 'cnt02', 'cnt03']
df
```

```python
df = df.reset_index()
df.columns = ['cnt01', 'cnt02', 'cnt03']
from matplotlib import rc
# y-axis in bold
rc('font', weight='bold')
 
# Values of each group
bars1 = df['cnt01']
bars2 = df['cnt02']
bars3 = df['cnt03']
 
# Heights of bars1 + bars2
bars = np.add(bars1, bars2).tolist()
 
# The position of the bars on the x-axis
r = np.arange(len(bars1))
 
# Names of group and bar width
names = df['date']
barWidth = 0.5

plt.figure(figsize=(18,5))
# Create brown bars
plt.bar(r, bars1, color='blue', edgecolor='white', width=barWidth, label='Masstricht')
# Create green bars (middle), on top of the first ones
plt.bar(r, bars2, bottom=bars1, color='red', edgecolor='white', width=barWidth, label='Rijswijk')
# Create green bars (top)
plt.bar(r, bars3, bottom=bars, color='green', edgecolor='white', width=barWidth, label='Zuidbroek')
 
# Custom X axis
plt.xticks(r, names, fontweight='bold', rotation=90)
plt.xlabel("Date")

plt.ylabel("Number of icesat2 points")

plt.legend()

# plt.savefig('for_report/date.pdf') # ,dpi=600
# Show graphic
plt.show()
```
